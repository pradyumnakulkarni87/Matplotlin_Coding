# Line plot

```python
# X: independant
days = [1,2,3,4,5,6,7]

#Y: dependant
cases = [10,50,5,67,12,30,10]
```

```python
plt.plot(cases)
plt.grid()
```

```python
plt.plot(cases)
plt.xticks(ticks=range(0,7),labels=['d1','d2','d3','d4','d5','d6','d7'])
plt.grid()
```

```python
# use both cases and days
plt.plot(days,cases)
```

```python
#set xlabel ylabel and title

plt.plot(days,cases)
plt.xlabel('Days')
plt.ylabel('Cases')
plt.title('Covid19 Weekly cases')
```

```python
#change color, line style and line width

plt.plot(days,cases,color='red')
#colors= red,yellow, black, cyan, magenta,green
plt.xlabel('Days')
plt.ylabel('Cases')
plt.title('Covid19 Weekly cases')
```

```python
#change color, line style and line width
#instead of default color i want customized color
plt.plot(days,cases,color='#8f1ad0',linestyle='-.',lw=4)
#colors= red,yellow, black, cyan, magenta,green
plt.xlabel('Days')
plt.ylabel('Cases')
plt.title('Covid19 Weekly cases')
```

```python
#label the line
plt.plot(days,cases,color='#8f1ad0',linestyle='-.',lw=4,label='Covid19 cases')
#colors= red,yellow, black, cyan, magenta,green
plt.xlabel('Days')
plt.ylabel('Cases')
plt.legend()
plt.title('Covid19 Weekly cases')
```

```python
#add markers to highlight intersection part
plt.plot(days,cases,color='#8f1ad0',lw=2,label='Covid19 cases',marker='D')
#colors= red,yellow, black, cyan, magenta,green
plt.xlabel('Days')
plt.ylabel('Cases')
plt.legend()
plt.title('Covid19 Weekly cases')
```

```python
# linear
plt.subplot(221)
plt.plot(days, cases)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(days, cases)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(days, cases)
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(days, cases)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
