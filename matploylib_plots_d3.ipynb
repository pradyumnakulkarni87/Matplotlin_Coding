# Pie plot

### it gives %wise distribustion of data

```python
#share market example
company = ['nifty','sensex','reliance','infosys']
```

```python
share = [98,45,70,33]
```

#### out of 100% what is the share of each company

```python
plt.pie(share)
plt.show()
```

```python
#add the labels of a company
plt.pie(share,labels=company)
plt.show()
```

```python
#display the % distribution use autoct
plt.pie(share,labels=company,autopct='%f%%')
plt.show()
```

```python
#upto 2 decimal places i want the %
plt.pie(share,labels=company,autopct='%.2f%%')
plt.show()
```

```python
plt.pie(share,labels=company,autopct='%.f%%')
plt.show()
```

```python
#colors
plt.pie(share,labels=company,autopct='%.f%%',colors=['y','c','m','g'])
plt.show()
```

```python
#change the position of labels and %
plt.pie(share,labels=company,autopct='%.f%%',colors=['y','c','m','g'],
       pctdistance=1.1,labeldistance=0.45)
plt.show()
```

# Convert data in Series

```python
import pandas as pd
s = pd.Series(data=share,index=company,name='Share market data')
s
```

```python
plt.pie(s.values,labels=s.index)
plt.title('Share market data')
plt.show()
```

```python
s.plot.pie(labeldistance=1.4)
plt.title('Share market data')
```

```python
import numpy as np
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.1)
  
# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey")
  
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
  
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
  
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(share, 
                                  autopct = lambda pct: func(pct, share),
                                  explode = explode, 
                                  labels = company,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
  
# Adding legend
ax.legend(wedges, company,
          title ="Share market data",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
  
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")
  
# show plot
plt.show()
```
