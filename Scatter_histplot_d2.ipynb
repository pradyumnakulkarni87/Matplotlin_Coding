# Bar plot

```python
states=['MAH','KE','MP','AP']
d2020 = [65,79,100,56]
```

```python
plt.bar(states,d2020)
```

```python
d2021 = [35,120,50,200]
plt.bar(states,d2021)
```

## Combine data of 2020 and 2021

```python
#2020
plt.bar(states,d2020,label='2020')

#2021
plt.bar(states,d2021,label='2021')
plt.legend()
```

# Seperate two bar plots

```python
#2020
plt.bar(states,d2020,label='2020')

#2021
plt.bar(states,d2021,label='2021',width=0.4,align='edge')
plt.legend()
```

```python
fig = plt.figure(figsize=(8,4))
plt.subplot(121)#(row,column,no)
plt.bar(states,d2020,label='20')
plt.legend()
plt.title('2020 data')

plt.subplot(122)
plt.bar(states,d2021,label='21',color='y')
plt.legend()
plt.title('2021 Data')

fig.suptitle("Combined plot for 2020 and 2021", fontsize=16)
```

## histogram

```python
sugar_level = [90,80,300,200,220,150,170,90,89,70,60,79,500,450,355,440,410,200,260,270,280,290]
```

## Histogram is used to show rangewise distribustion of data

```python
plt.hist(sugar_level)
```

```python
#low (50-80)
#Normal (80-120)
# High (120>)
#[50,80,120,250,500]
```

```python
plt.hist(sugar_level,bins=[50,80,120,250,500])
```

```python
plt.hist(sugar_level,bins=[50,80,120,250,500],label='Day1')
plt.hist(sugar_level*2,bins=[50,80,120,250,500],alpha=.3,label='Day2')
plt.legend()
```

```python
plt.hist(sugar_level,bins=[50,80,120,250,500],label='Day1',histtype='step',orientation='horizontal')
plt.hist(sugar_level*2,bins=[50,80,120,250,500],alpha=.3,label='Day2',orientation='horizontal')
plt.legend()
```

```python
plt.hist(sugar_level*2,bins=[50,80,120,250,500],alpha=.3,label='Day2',orientation='horizontal',histtype='barstacked',rwidth=.95)
```
